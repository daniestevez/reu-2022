```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
phase = np.fromfile('phase_measurement_10sps.f32', 'float32')
t = np.arange(phase.size) * 0.1
plt.plot(t, phase)
plt.title('Phase versus time')
plt.xlabel('Time (s)')
plt.ylabel('Phase (rad)');
```

```python
phase_unwrap = np.unwrap(phase)
plt.plot(t, phase_unwrap)
plt.title('Unwrapped phase versus time')
plt.xlabel('Time (s)')
plt.ylabel('Phase (rad)');
```

```python
n = 100  # 10 seconds frequency measurement interval
t_frequency = t[::n]
frequency = np.diff(phase_unwrap[::n]) / (2*np.pi * np.diff(t_frequency))
plt.plot(t_frequency[:-1], frequency)
plt.title('Frequency versus time')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)');
```
